```python
#ARBOL 1
import random
#esta funcion nos permite visualizar el arbol
def print_tree(tree, indent=""):
    if isinstance(tree, tuple):
        feature, value, left, right = tree
        print(f"{indent}{feature} = {value}")
        print_tree(left, indent + "  ")
        print_tree(right, indent + "  ")
    else:
        print(f"{indent}Clase: {tree}")

def shuffle_data(data):
    random.shuffle(data)
#aqui calculamos la impureza de gini con respecto a todas mis clasificaciones
def calculate_gini(data):
    total_samples = len(data)
    if total_samples == 0:
        return 0

    gini = 1
    class_counts = {}
    for row in data:
        label = row["nom_animal"]
        class_counts[label] = class_counts.get(label, 0) + 1

    for label in class_counts:
        probability = class_counts[label] / total_samples
        gini -= probability ** 2

    return gini
#aqui dividimos lo sa datos
def split_data(data, feature, value):
    left = [row for row in data if row[feature] == value]
    right = [row for row in data if row[feature] != value]
    return left, right
#aqui me toma la mejor division para hacer el arbol y predecir la proxima opcion
def find_best_split(data, features):
    best_gini = 1
    best_feature = None
    best_value = None

    for feature in features:
        values = set(row[feature] for row in data)
        for value in values:
            left, right = split_data(data, feature, value)
            gini = (len(left) / len(data)) * calculate_gini(left) + (len(right) / len(data)) * calculate_gini(right)
            if gini < best_gini:
                best_gini = gini
                best_feature = feature
                best_value = value

    print("Mejor Split:")
    print("Característica:", best_feature)
    print("Valor:", best_value)
    print("Gini:", best_gini)

    return best_feature, best_value
#aqui construye el arbol
def build_tree(data, features):
    if not data:
        return None
    if len(set(row["nom_animal"] for row in data)) == 1:
        return data[0]["nom_animal"]
    if not features:
        return max(set(row["nom_animal"] for row in data), key=lambda x: [row["nom_animal"] for row in data].count(x))

    best_feature, best_value = find_best_split(data, features)
    left_data, right_data = split_data(data, best_feature, best_value)
    remaining_features = [f for f in features if f != best_feature]

    left_subtree = build_tree(left_data, remaining_features)
    right_subtree = build_tree(right_data, remaining_features)

    return (best_feature, best_value, left_subtree, right_subtree)
#en estas dos funciones nos predice el animal insertado
def predict(tree, samples):
    predictions = [_predict(tree, sample) for sample in samples]
    return predictions

def _predict(tree, sample):
    if not isinstance(tree, tuple):
        return tree
    feature, value, left, right = tree
    return _predict(left, sample) if sample[feature] == value else _predict(right, sample)

shuffle_data(animales)
#set de entrenamiento



total_datos = len(animales)
por_train = 0.7
tam_data_train = int(total_datos * por_train)

data_test = animales[tam_data_train:]

tree = build_tree(animales, ["clase", "dieta", "extremidades", "Tipo_de_pelaje", "sensacion_de_pelaje", "hogar", "tamano_animal", "color_caracteristico","reproduccion","actitud"])



import random

# set de entrenamiento
shuffle_data(animales)


por_train = 0.7
total_datos = len(animales)
tam_data_train = int(total_datos * por_train)

# guardamos los dos set en entrenamiento y prueba
data_train = animales
data_test = animales[tam_data_train:]

#hacemos la prediccion del animal
input = data_test

predictions = predict(tree, input)
print("-----------------------------------------------------------")
print("Tu animal pensado es:", predictions)

def calculate_accuracy(predictions, actual_labels):
    correct = sum(1 for prediction, actual in zip(predictions, actual_labels) if prediction == actual)
    accuracy = (correct / len(actual_labels)) * 100
    return accuracy
print("-----------------------------------------------------------")
print("Árbol de decisión:")
print_tree(tree)


actual_labels = [sample["nom_animal"] for sample in data_test]

accuracy = calculate_accuracy(predictions, actual_labels)

```

```python
#cual es la clase mas comun

def clase_mas_comun(data):
    clas = [fila["nom_animal"] for fila in data]
    return max(set(clas), key=clas.count)
```

```python
def calcular_gini(data):
    if not data:
        return 0.0

    clas = [fila["nom_animal"] for fila in data]
    tot = len(clas)
    impu = 1.0

    for clase in set(clas):
        prob = clas.count(clase) / tot
        impu -= prob ** 2
    return impu
```

```python
def find_best_split(data, features):
    best_gini = 1
    best_feature = None
    best_value = None

    for feature in features:
        values = set(row[feature] for row in data)
        for value in values:
            left, right = split_data(data, feature, value)
            gini = (len(left) / len(data)) * calculate_gini(left) + (len(right) / len(data)) * calculate_gini(right)
            if gini < best_gini:
                best_gini = gini
                best_feature = feature
                best_value = value

    print("Mejor Split:")
    print("Característica:", best_feature)
    print("Valor:", best_value)
    print("Gini:", best_gini)

    return best_feature, best_value
```

```python
def predecir_animal(animal, arbol):
    if isinstance(arbol, dict):
        atributo = arbol["atributo"]
        valor = arbol["valor"]
        if animal.get(atributo) == valor:
            return predecir_animal(animal, arbol["izquierdo"])
        else:
            return predecir_animal(animal, arbol["derecho"])
    else:
        return arbol
```

```python
def construir_arbol(data):
    if not data:
        return clase_mas_comun(data)

    clases = set(fila["nom_animal"] for fila in data)


    if len(clases) == 1:
        return clases.pop()


    if len(data[0]) == 1:
        return clase_mas_comun(data)


    mejor_gini = 1.0
    mejor_division = None

    for atributo in data[0]:
        if atributo != "nom_animal":
            valores = set(fila[atributo] for fila in data)
            for valor in valores:
                conjunto_izquierda = [fila for fila in data if fila[atributo] == valor]
                conjunto_derecha = [fila for fila in data if fila[atributo] != valor]

                gini_izquierda = calcular_gini(conjunto_izquierda)
                gini_derecha = calcular_gini(conjunto_derecha)

                gini_total = (len(conjunto_izquierda) / len(data)) * gini_izquierda + (len(conjunto_derecha) / len(data)) * gini_derecha

                if gini_total < mejor_gini:
                    mejor_gini = gini_total
                    mejor_division = (atributo, valor, conjunto_izquierda, conjunto_derecha)

    if mejor_gini == 1.0:
        return clase_mas_comun(data)
    atributo, valor, conjunto_izquierda, conjunto_derecha = mejor_division
    subarbol_izquierdo = construir_arbol(conjunto_izquierda)
    subarbol_derecho = construir_arbol(conjunto_derecha)

    return {"atributo": atributo, "valor": valor, "izquierdo": subarbol_izquierdo, "derecho": subarbol_derecho}
```
